```python
# Stack Min: How would you design a stack which, in addition to push and pop, has a function min which returns the minimum element? Push, pop and min should all operate in 0(1) time.

class StackMin:
    def __init__(self):
        self.stack = []
        self.min = []

    def push(self, value):
        self.stack.append(value)
        if len(self.min) == 0:
            self.min.append(value)
        else:
            if value <= self.min[-1]:
                self.min.append(value)

    def pop(self):
        if len(self.stack) == 0:
            return None
        else:
            value = self.stack.pop()
            if value == self.min[-1]:
                self.min.pop()
            return value

    def get_min(self):
        if len(self.min) == 0:
            return None
        else:
            return self.min[-1]
```

```python
# Stack of Plates: Imagine a (literal) stack of plates. If the stack gets too high, it might topple. Therefore, in real life, we would likely start a new stack when the previous stack exceeds some threshold. Implement a data structure SetOfStacks that mimics this. SetOfStacks should be composed of several stacks and should create a new stack once the previous one exceeds capacity. SetOfStacks. push () and SetOfStacks. pop () should behave identically to a single stack (that is, pop ( ) should return the same values as it would if there were just a single stack).

# FOLLOW UP; Implement a function popAt ( int index) which performs a pop operation on a specific sub-stack.

class SetOfStacks:
    def __init__(self, threshold):
        self.threshold = threshold
        self.stacks = [[]]

    def push(self, value):
        if len(self.stacks[-1]) < self.threshold:
            self.stacks[-1].append(value)
        else:
            self.stacks.append([value])

    def pop(self):
        # If the last stack in the list is empty, return None
        if len(self.stacks[-1]) == 0:
            return None
        # Otherwise, pop the last element off of the stack
        else:
            value = self.stacks[-1].pop()
            # If the stack is now empty, remove it from the list
            if len(self.stacks[-1]) == 0:
                self.stacks.pop()
            # Return the value
            return value

    def pop_at(self, index):
        # If the stack at the given index is empty, return None
        if len(self.stacks[index]) == 0:
            return None
        # Otherwise, pop the last element off of the stack at the given index
        else:
            value = self.stacks[index].pop()
            # If the stack is now empty, remove it from the list
            if len(self.stacks[index]) == 0:
                self.stacks.pop(index)
            # Return the value
            return value
```

```python
# Sort Stack: Write a program to sort a stack such that the smallest items are on the top. You can use an additional temporary stack, but you may not copy the elements into any other data structure (such as an array). The stack supports the following operations: push, pop, peek, and isEmpty.

def sort_stack(stack):
    # Create a temporary stack
    temp_stack = []
    # While the stack is not empty
    while len(stack) > 0:
        # Pop the top element off of the stack
        temp = stack.pop()
        # While the temporary stack is not empty and the top element is greater than the temporary stack's top element
        while len(temp_stack) > 0 and temp > temp_stack[-1]:
            # Pop the temporary stack's top element and push it onto the original stack
            stack.append(temp_stack.pop())
        # Push the temporary variable onto the temporary stack
        temp_stack.append(temp)
    # Return the temporary stack
    return temp_stack
```
